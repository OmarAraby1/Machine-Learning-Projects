# 1. Load and Explore the Data

```python
# import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the training dataset and explore it
train = pd.read_csv('./datasets/train.csv')
train.head()
```

```python
train.describe()
```

```python
train.info()
```

```python
# Check for missing values
print(train.isnull().sum())
```

## Deal with Null Values

- As there are many Null values in the "Cabin" column, we might just drop it.

```python
# Impute missing Age values
# Create a function to fill missing ages
def impute_age(row):
    if pd.isnull(row['Age']):
        return age_medians.loc[row['Survived'], row['Pclass'], row['Sex']]
    else:
        return row['Age']

# Calculate the median age for each Survived, Pclass and Sex group
age_medians = train.groupby(['Survived','Pclass', 'Sex'])['Age'].median()

# Apply the function to fill missing Age values
train['Age'] = train.apply(impute_age, axis=1)

# Impute missing Embarked values with the most frequent value
most_frequent_embarked = train['Embarked'].mode()[0]
train['Embarked'].fillna(most_frequent_embarked, inplace=True)

# Verify that there are no more missing values
print(train.isnull().sum())
```

## Explore Outliers

```python
# Create a set of subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Box plot using matplotlib
axes[0].boxplot([train['Age']])
axes[0].set_xticklabels(['Age'])
axes[1].boxplot([train['Fare']])
axes[1].set_xticklabels(['Fare'])

# Display the figures
plt.tight_layout()
plt.show()
```

```python
def identify_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Identify outliers in Age and Fare columns
outliers_age = identify_outliers(train, 'Age')
outliers_fare = identify_outliers(train, 'Fare')

print(f"Outliers in Age:\n{outliers_age}")
print(f"Outliers in Fare:\n{outliers_fare}")
```

# Feature Engineering

```python
# Bin ages into categories
train['AgeBin'] = pd.cut(train['Age'], bins=[0, 12, 20, 40, 60, 80], labels=['Child', 'Teen', 'Adult', 'Middle-aged', 'Senior'])

# Create family size feature
train['FamilySize'] = train['SibSp'] + train['Parch'] + 1
```

## Encode Categories

```python
from sklearn.preprocessing import OneHotEncoder

# Encode categorical features
categorical_features = ['Sex', 'Embarked', 'AgeBin']
train = pd.get_dummies(train, columns=categorical_features)
```

# Split Data

```python
from sklearn.model_selection import train_test_split

# Define features and target variable
X = train.drop(columns=['Survived', 'Name', 'Cabin', 'Ticket', 'PassengerId'])
y = train['Survived']

# Split the data
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Build a Base Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Train the model
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

# Make predictions
y_pred_logreg = logreg.predict(X_val)

# Evaluate the model
print('Logistic Regression Accuracy:', accuracy_score(y_val, y_pred_logreg))
print(confusion_matrix(y_val, y_pred_logreg))
print(classification_report(y_val, y_pred_logreg))
```

# Build a Base Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier

# Train the model
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)

# Make predictions
y_pred_dtree = dtree.predict(X_val)

# Evaluate the model
print('Decision Tree Accuracy:', accuracy_score(y_val, y_pred_dtree))
print(confusion_matrix(y_val, y_pred_dtree))
print(classification_report(y_val, y_pred_dtree))
```

# Build a Base Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier

# Train the model
rforest = RandomForestClassifier()
rforest.fit(X_train, y_train)

# Make predictions
y_pred_rforest = rforest.predict(X_val)

# Evaluate the model
print('Random Forest Accuracy:', accuracy_score(y_val, y_pred_rforest))
print(confusion_matrix(y_val, y_pred_rforest))
print(classification_report(y_val, y_pred_rforest))
```

# Tune the Random Forest Model

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rforest, param_grid=param_grid, cv=10, n_jobs=-1, verbose=2)

# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Best parameters
print(grid_search.best_params_)
```

# Test Our Tuned Paramteres

```python
# Train the model with the best parameters
best_rforest = grid_search.best_estimator_
best_rforest.fit(X_train, y_train)

# Make predictions
y_pred_best_rforest = best_rforest.predict(X_val)

# Evaluate the model
print('Best Random Forest Accuracy:', accuracy_score(y_val, y_pred_best_rforest))
print(confusion_matrix(y_val, y_pred_best_rforest))
print(classification_report(y_val, y_pred_best_rforest))
```

 - The Recall Value has increase which means we increased the T.P rate

```python
# Define a function to plot comparisons
def plot_comparison(feature, title):
    plt.figure(figsize=(14, 6))

    # Plot for the train set
    plt.subplot(1, 2, 1)
    sns.countplot(data=train, x=feature, hue='Survived')
    plt.title(f'Train Set - {title}')
    
    # Plot for the test set
    plt.subplot(1, 2, 2)
    sns.countplot(data=test, x=feature, hue='Survived')
    plt.title(f'Test Set - {title}')
    
    plt.show()

# Plot comparisons for selected features
features_to_compare = ['Pclass', 'Sex', 'AgeBin', 'FamilySize']

for feature in features_to_compare:
    plot_comparison(feature, feature)
```

```python
# Define the order of categories for Embarked
embarked_order = ['S', 'C', 'Q']

# Plot comparisons for Embarked feature with specified order
plt.figure(figsize=(14, 6))

# Plot for the train set
plt.subplot(1, 2, 1)
sns.countplot(data=train, x='Embarked', hue='Survived', order=embarked_order)
plt.title('Train Set - Embarked')

# Plot for the test set
plt.subplot(1, 2, 2)
sns.countplot(data=test, x='Embarked', hue='Survived', order=embarked_order)
plt.title('Test Set - Embarked')

plt.show()
```

```python
def plot_distribution(feature, title):
    plt.figure(figsize=(14, 6))

    # Distribution for the train set
    plt.subplot(1, 2, 1)
    sns.histplot(data=train, x=feature, hue='Survived', kde=True)
    plt.title(f'Train Set - {title}')
    
    # Distribution for the test set
    plt.subplot(1, 2, 2)
    sns.histplot(data=test, x=feature, hue='Survived', kde=True)
    plt.title(f'Test Set - {title}')
    
    plt.show()

# Plot distribution for continuous features
continuous_features_to_compare = ['Age', 'Fare']

for feature in continuous_features_to_compare:
    plot_distribution(feature, feature)
```
